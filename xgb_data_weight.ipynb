```python
train = train[col]
```

```python
# SUBCLASS 가 범주형이기 때문에 LabelEncoder 사용
le_subclass = LabelEncoder()
train['SUBCLASS'] = le_subclass.fit_transform(train['SUBCLASS'])

# 변환된 레이블 확인
for i, label in enumerate(le_subclass.classes_):
    print(f"원래 레이블: {label}, 변환된 숫자: {i}")
```

```python
## x 의 경우도 범주형으로 구성되어 있어, 알맞은 인코딩 필요
X = train.drop(columns=['SUBCLASS', 'ID'])
y_subclass = train['SUBCLASS']

categorical_columns = X.select_dtypes(include=['object', 'category']).columns
ordinal_encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
X_encoded = X.copy()
X_encoded[categorical_columns] = ordinal_encoder.fit_transform(X[categorical_columns])
```

```python
# 학습 및 테스트 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y_subclass, test_size=0.1, stratify=y_subclass, random_state=42)
```

```python
# 각 클래스별 샘플 수 계산
class_counts = np.bincount(y_train)

# 가중치 계산 (각 클래스 빈도의 역수로 설정)
weights = 1.0 / class_counts
sample_weights = np.array([weights[label] for label in y_train])

# XGBoost DMatrix 생성 (가중치 포함)
dtrain = xgb.DMatrix(X_train, label=y_train, weight=sample_weights)
dtest = xgb.DMatrix(X_test, label=y_test)
```

```python
# DMatrix 생성
dtrain = xgb.DMatrix(X_train, label=y_train)
dtest = xgb.DMatrix(X_test, label=y_test)

# 파라미터 설정
params = {
    'objective': 'multi:softmax',  # 다중 클래스 확률 예측
    'num_class': 26,                # 클래스 수 지정
    'eta': 0.1,                     # 학습률 (learning_rate)
    'max_depth': 12,                # 최대 깊이
    'colsample_bytree': 0.7,        # 각 트리에 사용하는 열의 비율
    'seed': 42,                     # random_state
    'eval_metric': 'mlogloss'       # 평가 메트릭
}
def f1_eval(preds, dtrain):
    labels = dtrain.get_label()
    preds = np.argmax(preds, axis=1)  # 예측 확률 중 가장 높은 것을 선택하여 클래스 예측
    f1 = f1_score(labels, preds, average='macro')  # 다중 클래스 F1 Score (macro average)
    return 'f1', f1
```

```python
# 모델 학습
num_rounds = 600  # n_estimators와 동일
evals = [(dtrain, 'train'), (dtest, 'eval')]
model = xgb.train(params, dtrain, num_boost_round=num_rounds, evals=evals, feval=f1_eval)
```

```python
# 예측
y_pred = model.predict(dtest)

# 정확도 계산 (예시)
accuracy = np.mean(y_pred == y_test)
print(f"Accuracy: {accuracy * 100:.2f}%")
```
