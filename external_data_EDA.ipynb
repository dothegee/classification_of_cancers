### SUBCLASS 33개 -> 26개로 묶기
- KIRP, KICH → KIPAN으로 통합.
- GBM → GBMLGG로 통합.
- STAD, ESCA → STES로 통합.
- CHOL, MESO, UCS, UVM 제거.

```python
external_train['SUBCLASS'] = external_train['SUBCLASS'].replace({
    'KIRP': 'KIPAN', 'KICH': 'KIPAN', 
    'GBM': 'GBMLGG', 
    'STAD': 'STES', 'ESCA': 'STES',
})
```

```python
values_to_remove = ['CHOL', 'MESO', 'UCS', 'UVM', 'READ']

# 제거할 값들이 포함된 행 제거
external_train = external_train[~external_train['SUBCLASS'].isin(values_to_remove)]

# 결과 출력

external_train['SUBCLASS'].unique()
```

```python
check_external_train = external_train.iloc[:,2:]
```

```python
import re
# 'wt'가 아닌 값만 리스트로 모으기
non_wt_values = check_external_train.applymap(lambda x: x if x != 'WT' else 'WT')

# 'wt'가 아닌 값만 추출하여 리스트로 변환
non_wt_list = non_wt_values[non_wt_values != 'WT'].stack().tolist()
non_wt = []
for x in non_wt_list:
    temp = x.split()
    for y in temp:
        non_wt.append(y)

list_a=[]
for data in non_wt: 
    list_a.append(data.split(',')[0])

data_without_numbers = [re.sub(r'[A-Z0-9]', '', data) for data in list_a]
result = [item for item in data_without_numbers if item != '']
set(result)
```
